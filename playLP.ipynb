```python
import pyomo

import pyomo.environ as pyo

D = {'Krunchies','Crispies'}

N = {'Thiamine','Niacin','Calories'}

h = {'Thiamine': 1, 'Niacin' : 5, 'Calories' : 400}

c = { 'Krunchies' : 3.8, 'Crispies' : 4.2}

a = {'Krunchies':{'Thiamine': 0.1, 'Niacin': 1.0, 'Calories': 110 },\
     'Crispies':{'Thiamine': 0.25,'Niacin': 0.25, 'Calories': 120.0}}



model = pyo.ConcreteModel()

model.x = pyo.Var(D, within=pyo.NonNegativeReals)

model.sufficient = pyo.Constraint( N, rule = lambda m,n:\
                                  sum(a[d][n] * model.x[d]\
                                  for d in D) >= h[n])

model.cost = pyo.Objective(expr = sum(c[d]*model.x[d] for d in D),\
                           sense = pyo.minimize)

solver = pyomo.opt.SolverFactory ('cbc', executable='cbc')

results = solver.solve(model)

for d in D:
  print (d, model.x[d].value)
print ('')
print("Objective function (cost): " + str(model.cost()))
```
